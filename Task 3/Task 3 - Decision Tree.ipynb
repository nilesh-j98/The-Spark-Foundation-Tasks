### For the given ‘Iris’ dataset, create the Decision Tree classifier and visualize it graphically.

```python
#Importing the libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier 
from sklearn.model_selection import train_test_split
from sklearn import metrics #For accuracy calculation
```

```python
#Loading the dataset
col_names = ['Id','SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm','Species']
data = pd.read_csv("Iris.csv", header=None, names=col_names)
```

```python
#Viewing the Dataset
data.head(10)
```

## Splitting the dataset in features and target variables 

```python
feature_cols = ["SepalLengthCm","SepalWidthCm","PetalLengthCm","PetalWidthCm"]
X = data[feature_cols] #Features
X = X[1:]
Y = data.Species
Y = Y[1:]
```

## Splitting the dataset into training and testing set 

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.3, random_state = 1) #30% training and 70% test
```

## Building Decision Tree Model 

```python
#DecisionTreeClassifier object
clf = DecisionTreeClassifier()

#Train the DecisionTreeClassifier
clf = clf.fit(X_train,Y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

## Evaluating the Model

```python
print("Accuracy = ", metrics.accuracy_score(Y_test, y_pred))
```

## Importing the Data Visualizing Packages 

```python
import matplotlib.pyplot as plt
from sklearn import tree
```

```python
fig = plt.figure(figsize=(25,20))
_ = tree.plot_tree(clf,
                   feature_names = feature_cols,
                   class_names = "Species",
                   filled = True)
```
